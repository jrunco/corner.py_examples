```python
import numpy as np
import matplotlib.pyplot as plt
import corner
```

```python
data1x = np.random.normal(100.0, 10.0, 10000)
data1y = np.random.normal(100.0, 10.0, 10000)
data1 = np.vstack((data1x, data1y)).T
```

```python
data2x = np.random.normal(200.0, 15.0, 10000)
data2y = np.random.normal(200.0, 15.0, 10000)
data2 = np.vstack((data2x, data2y)).T
```

```python
corner.corner(data1, color='b')
corner.corner(data2)
plt.show()
```

```python
# plot both distributions together
example_fig = corner.corner(data1, bins=20, show_titles=True, color='b')
corner.corner(data2, bins=50, range=[(np.min(data1x),np.max(data2x)), (np.min(data1y),np.max(data2y))], 
              labels=["$x-data$", "y-data"], show_titles=True, fig=example_fig)
plt.show()
```

```python
# useful things to include
example_fig2 = corner.corner(data1, bins=20, fill_contours=True, quantiles=[0.16, 0.5, 0.84], color='b')
corner.corner(data2, bins=50, range=[(np.min(data1x),np.max(data2x)), (np.min(data1y),np.max(data2y))], 
              labels=["$x-data$", "$y-data$"], fill_contours=True, quantiles=[0.16, 0.5, 0.84], 
              hist_kwargs={'hatch':'\\\\', 'alpha':0.3, 'fill':True}, fig=example_fig2)
plt.show()
```
